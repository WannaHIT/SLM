```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data = np.array([
    [2008, 6000],
    [2009, 8000],
    [2010, 9000],
    [2011, 11000],
    [2012, 12000],
])
```

```python
# init params k, b
k = np.random.random()
b = np.random.random()
print(k)
print(b)
```

```python
data.T
```

```python
# 数据解包
X, Y = data.T
print(X)
print(Y)
```

```python
xalpha = X.mean()
xbeta = X.std()
```

```python
yalpha = Y.mean()
ybeta = Y.std()
```

```python
# 标准化
X = (X - xalpha) / xbeta
Y = (Y - yalpha) / ybeta
```

```python
print(X)
print(Y)
```

```python
niter = 0
```

```python
while True:
    niter += 1
    P = k * X + b
    L = np.mean(0.5 * (P - Y)**2)
    
    raw_x = X * xbeta + xalpha
    raw_y = Y * ybeta + yalpha
    raw_p = P * ybeta + yalpha
    
    # Clear the current axes.
    plt.cla()
    plt.plot(raw_x, raw_y, "b*")
    plt.plot(raw_x, raw_p, "r-")
    plt.pause(0.01)
    
    print(f"{niter}. Loss: {L:.5f}, k={k:.5f}, b={b:.5f}")
    
#     if L < 1e-4:
#         print(f"Done , {niter}")
#         break
        
    if niter >= 100:
        print(f"Done , {niter}")
        break
        
    deltak = np.mean((P - Y) * X)
    deltab = np.mean(P - Y)
    
    lr = 0.1
    k = k - lr*deltak
    b = b - lr*deltab
```

```python
predict_year = 2013
normalize_year = (predict_year - xalpha) / xbeta
normalize_pred = k * normalize_year + b
real_pred = normalize_pred * ybeta + yalpha
real_pred
```
